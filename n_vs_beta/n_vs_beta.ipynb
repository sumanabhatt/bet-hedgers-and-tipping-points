```python
import matplotlib.pyplot as plt
import numpy as np

# Data
n = [0, 1, 2, 3, 4, 5, 6, 7]

SN1 = [111.85398175, 104.65733217, 85.40767596, 56.57057132, 29.06988374, 11.15691569, 3.02218045, 0.51948994] 

SN2 = [169.08988624, 158.21077635, 129.11126391, 85.5175647, 43.94474309, 16.86573322, 4.56864179, 0.78531211]

# Calculate SN2 - SN1
SN_diff = [(s2 - s1) for s2, s1 in zip(SN2, SN1)]

# Create the plot
plt.figure(figsize=(12, 8))
plt.plot(n, SN_diff, 'o-', color='purple', linewidth=3, alpha=0.7, markersize=10, label=r'$\beta^{(2)}_\text{crit}-\beta^{(1)}_\text{crit}$')

# Set the labels and title
plt.xlabel('n', fontsize=24)
plt.ylabel(r'$\beta^{(2)}_\text{crit}-\beta^{(1)}_\text{crit}$', fontsize=24)          

# Set y-axis ticks (adjust as needed based on your data)
plt.yticks(fontsize=20)

# Add legend
legend = plt.legend(
    fontsize=20,
    borderaxespad=0.1,
    frameon=True,            
    fancybox=True,       
    framealpha=1.0)

# Add grid
plt.grid(True, linestyle=':', alpha=0.5)

# Show the plot
plt.xticks(n, fontsize=20)  # Use the actual n values for x-ticks
plt.savefig('n_vs_delta_beta.jpg', dpi=300, bbox_inches='tight')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import FixedLocator

# Data
n = [0, 1, 2, 3, 4, 5, 6, 7]

SN1 = [111.85398175, 104.65733217, 85.40767596, 56.57057132, 29.06988374, 11.15691569, 3.02218045, 0.51948994] 

SN2 = [169.08988624, 158.21077635, 129.11126391, 85.5175647, 43.94474309, 16.86573322, 4.56864179, 0.78531211]

# Calculate SN2 - SN1
SN_diff = [(s2 - s1) for s2, s1 in zip(SN2, SN1)]


# Create the main plot
fig, ax1 = plt.subplots(figsize=(12, 8))

# Plot main data
ax1.plot(n, SN1, 'o-', color='blue', linewidth=3, alpha=0.7, markersize=10, label=r'$\beta^{(1)}_\text{crit}(n)$')
ax1.plot(n, SN2, 'o-', color='red', linewidth=3, alpha=0.7, markersize=10, label=r'$\beta^{(2)}_\text{crit}(n)$')

# Set labels and ticks for main plot
ax1.set_xlabel('n', fontsize=24)
ax1.set_ylabel(r'$\beta_\text{crit}(n)$', fontsize=24)
ax1.set_xlim(0, )
ax1.set_ylim(0.0,)
ax1.set_yticks([0, 40, 80, 120, 160],[0.0, 40, 80, 120, 160])
#ax1.set_xticks([0, 1, 2, 3, 4, 5, 5.5],['0', '1','2','3','4','5',''])
ax1.tick_params(axis='both', which='major', labelsize=20)

# Add legend to main plot
legend = ax1.legend(
    fontsize=20,
    borderaxespad=0.1,
    loc='lower left',
    bbox_to_anchor=(0.005, 0.01), 
    frameon=True,            
    fancybox=True,       
    framealpha=1.0
)

# Add grid to main plot
ax1.grid(True, linestyle=':', alpha=0.5)

# Create the inset axes
#ax2 = fig.add_axes([0.17, 0.16, 0.45, 0.45])
ax2 = fig.add_axes([0.575, 0.56, 0.40, 0.40])

# Plot data in the inset
ax2.plot(n, SN_diff, 'o-', color='GREEN', linewidth=2, alpha=0.7, markersize=6)

# Set labels and ticks for inset
ax2.set_xlabel('n', fontsize=24)
ax2.set_ylabel(r'$\beta^{(2)}_\text{crit}(n)-\beta^{(1)}_\text{crit}(n)$', fontsize=24)
#ax2.set_xticks(range(7))
ax2.set_xlim(0, )
ax2.set_ylim(0.0,)
#ax2.set_xticklabels(['0', '1', '2', '3', '4', '5', '6','7'])
ax2.set_yticks([0, 10, 20, 30, 40, 50, 60])
ax2.set_yticklabels(['0.0', '', '20', '', '40', '', '60'])
ax2.tick_params(axis='both', which='major', labelsize=20)

# Add legend to inset
#ax2.legend(fontsize=20, borderaxespad=0.1, frameon=True, fancybox=True, framealpha=1.0)

# Add grid to inset
ax2.grid(True, linestyle=':', alpha=0.5)

# Adjust layout and save
plt.tight_layout()
plt.savefig('n_vs_beta_with_inset.jpg', dpi=300, bbox_inches='tight')
plt.show()
```
