```python
import numpy as np
import matplotlib.pyplot as plt

def read_bifurcation_data(filename):
    return np.loadtxt(filename, delimiter=',', skiprows=1)

def plot_bifurcation_diagram(filenames0, filenames1, filenames2):
    plt.figure(figsize=(12, 8))
    
    # Plot first set of data (n=28)
    for i, filename in enumerate(filenames0):
        data = read_bifurcation_data(filename)
        betas, x_values, stability = data.T  # Transpose to get columns
        
        if i == 0:  # Main bifurcation diagram
            plt.scatter(betas[stability == 0], x_values[stability == 0], 
                s=6, c='orangered', alpha=0.7, label='Stable (n=27)')
            plt.plot(betas[stability == 1], x_values[stability == 1], 
                c='teal',linestyle='--', label='Unstable (n=27)', alpha=1, linewidth=2)
        else:  # Saddle-node bifurcations
            plt.scatter(betas[stability == 2], x_values[stability == 2], 
                s=40, c='green', alpha=1)
            
    # Plot first set of data (n=29)
    for i, filename in enumerate(filenames1):
        data = read_bifurcation_data(filename)
        betas, x_values, stability = data.T  # Transpose to get columns
        
        if i == 0:  # Main bifurcation diagram
            plt.scatter(betas[stability == 0], x_values[stability == 0], 
                s=6, c='red', alpha=0.7, label='Stable (n=28)')
            plt.plot(betas[stability == 1], x_values[stability == 1], 
                c='blue', linestyle='--', label='Unstable (n=28)', alpha=1, linewidth=2)
        else:  # Saddle-node bifurcations
            plt.scatter(betas[stability == 2], x_values[stability == 2], 
                s=40, c='green', alpha=1)

    # Plot second set of data (n=30)
    for i, filename in enumerate(filenames2):
        data = read_bifurcation_data(filename)
        betas, x_values, stability = data.T  # Transpose to get columns
        
        if i == 0:  # Main bifurcation diagram
            plt.scatter(betas[stability == 0], x_values[stability == 0], 
                s=6, c='darkred', alpha=0.7, label='Stable (n=29)')
            plt.plot(betas[stability == 1], x_values[stability == 1], 
                c='darkblue', linestyle='--', label='Unstable (n=29)', alpha=1, linewidth=2)
        else:  # Saddle-node bifurcations
            plt.scatter(betas[stability == 2], x_values[stability == 2], 
                s=40, c='green', alpha=1)

            
    plt.xlabel(r'${\beta}$', fontsize=24)
    plt.ylabel('Fixed Points', fontsize=24)
    legend = plt.legend(
    fontsize=16,
    loc=(0.7, 0.04),
    borderaxespad=0.1,
    frameon=True,            
    fancybox=True,       
    framealpha=1.0, markerscale=6)  
    plt.grid(True, linestyle=':', alpha=0.5)
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
    plt.ylim(0, 0.42)
    plt.xlim(0,70)
#    plt.xticks([0.0, 6.76628329, 10, 10.22857286, 13.18217054, 19.92724091, 20, 24.08477119,  30, 36.4084507, 40 ], 
#               ['0.0', r'$\beta_{crit}^{(1)}(30)$', '', r'$\beta_{crit}^{(2)}(30)$', r'$\beta_{crit}^{(1)}(29)$',   r'$\beta_{crit}^{(2)}(29)$', '', r'$\beta_{crit}^{(1)}(28)$', '30', r'$\beta_{crit}^{(2)}(28)$', '40'], rotation=45)
    plt.xticks([0.0, 10, 20, 30, 40, 50, 60, 70 ], 
               ['0.0', '10', '20', '30', '40','50','60','70'])
    plt.yticks([0.0, 0.05, 0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4],
               ['0.0','','0.1','','0.2','','0.3','','0.4'])
#    plt.text(35,0.398, 'n=27, 28, 29', fontsize=22, color='black', ha='center', va='center')
#    plt.text(21,0.06, 'F1', fontsize=24, color='black', ha='center', va='center')
#    plt.text(40,0.30, 'F2', fontsize=24, color='black', ha='center', va='center')
8
# List of files to plot
files_to_plot_0 = ['data0_n=28.txt', 'data1_n=28.txt', 'data2_n=28.txt']
files_to_plot_1 = ['data0_n=29.txt', 'data1_n=29.txt', 'data2_n=29.txt']
files_to_plot_2 = ['data0_n=30.txt', 'data1_n=30.txt', 'data2_n=30.txt']
#files_to_plot_3 = ['data0_n=31.txt', 'data1_n=31.txt', 'data2_n=31.txt']



# Plot the bifurcation diagram
plot_bifurcation_diagram(files_to_plot_0, files_to_plot_1, files_to_plot_2)
plt.savefig('comparison.jpg', dpi=300, bbox_inches='tight')
plt.show()
```
